# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# 1.Problemin Çözümünü Buraya Yazınız
import os

def vki_hesapla(kilo,boy):
    # kilo kg, boy m cinsinden olmalı
    # formul: kilo/boy**2
    if kilo > 635:
        kilo = kilo / 1000
    if boy > 5:
        boy = boy / 100
    vki = kilo / (boy ** 2)
    return vki

def kategori(vki):
    if vki < 18.5:
        return "zayıf"
    elif vki < 24.9:
        return "normal kilolu"
    elif vki < 29.9:
        return "fazla kilolu"
    elif vki < 39.9:
        return "obez"
    elif vki >= 39.9:
        return "morbid obez"
    else:
        print("bir hata oluştu!")
    

kisi_sayisi = int(input("islem yapilacak kisi sayisi: "))
kisi_list = []

for i in range(kisi_sayisi):
    ad_soyad = input(f"{i + 1}. kisinin isim soyisim: ")
    boy = float(input("boyunuz: "))
    kilo = float(input("kilonuz: "))

    vki = vki_hesapla(kilo, boy)
    kisi_kategori = kategori(vki)

    info = {"Ad Soyad: " : ad_soyad,
            "Kilo: " : kilo,
            "Boy: " : boy,
            "Vücut Kitle İndeksi: " : vki,
            "Kitlekse Göre Bulunulan Kategori: " : kisi_kategori}
    
    kisi_list.append(info)

with open("kitle_file.txt", "a", encoding="utf-8") as f:
    for i in kisi_list:
        f.write(f"Ad Soyad: {i["Ad Soyad: "]}\n")
        f.write(f"Kilo: {i["Kilo: "]}\n")
        f.write(f"Boy: {i["Boy: "]}\n")
        f.write(f"Vücut Kitle İndeksi: {i["Vücut Kitle İndeksi: "]}\n")
        f.write(f"Kitlekse Göre Bulunulan Kategori: {i["Kitlekse Göre Bulunulan Kategori: "]}\n")
        f.write("\n\n")


if os.path.exists("kitle_file.txt"):
    print("Bilgiler dosyaya kaydedilmiştir.")
```

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
# 2.Problemin Çözümünü Buraya Yazınız
# envanter sayısı ekle
import time
from datetime import datetime

def menu_gor():
    with open("menu.txt", "r", encoding="utf-8") as f:
        for urun in f:
            print(urun)
    
def satin_al():
    with open('menu.txt', 'r', encoding='utf-8') as file:
        lines = file.readlines()

    find = True
    urun = input("İstediğiniz ürünü giriniz: ").strip()
    while find:
        for line in lines:
            icecek, fiyat = line.strip().split(", ") 
            if urun == icecek:
                print(f"{icecek}: {fiyat} TL")
                find = False
                break  
     
    with open("alinanlar.txt", "r+", encoding="utf-8") as f:
        icerik = f.read()
        # f.write(icerik)
        f.write(f"Alınan ürün: {urun}\n")
        f.write(f"Fiyatı: {fiyat}\n")
        f.write(f"Tarih: {datetime.now()}\n")

    print("Satın alma işlemi tamamlanmıştır.")


def urun_ekle():
    islem_sayisi = int(input("Kaç ürün girmek istersiniz: "))
    urun_list = []
    for i in range(islem_sayisi):
        urun_adi = input(f"{i + 1}. ürünün adını giriniz: ")
        urun_fiyat = float(input("Ürünün fiyatını giriniz: "))

        urun_bilgi_list = {"Ad" : urun_adi, 
                           "Fiyat" : urun_fiyat}
        
        urun_list.append(urun_bilgi_list)

    with open("menu.txt", "a", encoding="utf-8") as f:
        for i in urun_list:
            f.write(f"{i["Ad"]}, {i["Fiyat"]}\n")

    print("Ürünler dosyaya eklenmiştir.")



while True:
    secim = input('''YetBar - Smoothie Bank\'a Hoş Geldiniz!!
                  Aşağıda yazan talimatlara uyarak istediğiniz işlemi gerçekleştirebilirsiniz: 
                 - Menüyü Görüntülemek için (1) tuşlayınız
                 - Satın Alma işlemi için (2) tuşlayınız
                 - Çıkış yapmak için (q) tuşlayınız
                 - Menüye Ürün eklemek için (3) tuşlayınız ''')
    
    if secim == "q":
        print("Çıkış yapılıyor...")
        break
    if secim == "1":
        print("Menüyü aşağıdan görüntülüyebilirsiniz: ")
        time.sleep(1)
        menu_gor()
    if secim == "2":
        print("Satın alma işleminizi gerçekleştirebilirsiniz: ")
        time.sleep(1)
        satin_al()
    if secim == "3":
        print("Menüye ürün girebilirsiniz: ")
        time.sleep(1)
        urun_ekle()



```
