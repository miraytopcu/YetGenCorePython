# Problem 1
```test.txt``` dosyasındaki satır sayısını bularak ekrana yazdırınız.



```
> Output
Number of lines in the file:  6

```



````python
#Sorunun Çözmünü Buraya Gerçekleştirebilirsiniz
def countLine(file_path):
    try:
        with open(file_path, "r", encoding="utf-8") as file:
            lines = file.readlines()
            number = len(lines)
            print("Number of lines in the file: ", number)
    except FileNotFoundError:
        print("The file doesnt exist.")

    file_path = 'test.txt'

    countLine(file_path)
````
