Çeşitli metin dosyalarından karakterleri ayıklamak ve bir listeye koymak için bir Python programı yazın. ```test.txt``` ve ```demo.txt``` dosyalarından faydalanabilirsiniz.

````python
#Sorunun Çözmünü Buraya Gerçekleştirebilirsiniz
# glob kütüphanesini projeye dahil et
import glob
# char_list isimli bir değişken oluşturup boş bir liste atayalım
char_list = []
# files_list isimli bir değişken oluşturup glob kütüphanesi ile txt uzantılı dosyaları bulalım
files_list = glob.glob("*.txt")
# for ile file_list'i gezelim ve with ile dosyaları açalım
for file_elem in files_list:
    with open(file_elem, "r") as f:
        # char_list listesine okuduğumuz dosyaları ekleyelim
        char_list.append(f.read())
# char_list listesini ekrana bastıralım
print(char_list)
````
