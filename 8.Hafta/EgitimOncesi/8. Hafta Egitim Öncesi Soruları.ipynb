# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
# 1.Problemin Çözümünü Buraya Yazınız.
def urunList():
    ad = input("Ürün adını giriniz: ")
    fiyat = float(input("Ürün fiyatını giriniz: "))
    with open("urunler.txt", "a", encoding = "utf-8") as file:
        file.write(f"{ad} , {fiyat}\n")
        print("ürün eklendi!")

urunList()
    
````

````python
# 2.Problemin Çözümünü Buraya Yazınız
# NEDEN ESKİ LİSTEYİ SİLMEYİP AYNI LİSTEYİ TEKRAR YAZDIRIYOR ?????
def degistirme():
    eski_kelime = input("urunler dosyasında bulunup değiştirmek istediğiniz urunu giriniz: ")
    yeni_kelime = input("yeni urunu giriniz: ")

    with open("urunler.txt", "r+", encoding="utf-8") as file:
        icerik = file.read()
        yeni_icerik = icerik.replace(eski_kelime, yeni_kelime)
        file.seek(0)
        file.write(yeni_icerik)
        # file.truncate()

        print(f"{eski_kelime} ifadesi {yeni_kelime} ifadesi ile değiştirildi!")

degistirme()
````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
# 3.Problemin Çözümünü Buraya Yazınız
class Vehicle():
    def __init__(self, name, speed, mileage):
        self.name = name
        self.speed = speed
        self.mileage = mileage
        print(f"Name: {name} Speed: {speed} Mileage: {mileage}")

class Bus(Vehicle):
    def __init__(self, name, speed, mileage):
        super().__init__(name, speed, mileage)

print("deneme for vehicle")
Vehicle("School Volvo", 180, 12)
print("deneme for Bus")
Bus("School Volvo", 180, 12)
````
