# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız
from datetime import datetime
import random

channels = ["TRT1", "Show", "Fox", "TV8"]
apps = ["Youtube", "Netflix", "BluTV", "Amazon Prime", "Disney+", "BeIN CONNECT"]
fav_channels = []

class remote_control():
    def __init__(self, channel = 1, power = "closed", volume = 0):
        self.channel = channel
        self.power = power
        self.volume = volume


    def tv_power(self):
        if self.power == "closed":
            print("TV is on.")
            self.power = "open"
        elif self.power == "open":
            print("TV is off.")
            self.power = "closed"

    def display_volume(self):
        if self.power == "closed":
            print("TV is off!")
        else:
            print(f"Volume Level: {self.volume}")

    def increase_volume(self):
        if self.power == "closed":
            print("TV is off!")
        else:
            try:
                control = int(input("How much do you want to increase the volume: "))
                if control < 0:
                    print("Value must be positive.")
                else:
                    self.volume += control
                    print(f"New Volume Level: {self.volume}")
            except ValueError as e:
                print(f"Error: {e}")

    def decrease_volume(self):
        if self.power == "closed":
            print("TV is off!")
        else:
            try:
                control = int(input("How much do you want to decrease the volume: "))
                if control < 0:
                    print("Value must be positive.")
                else:
                    self.volume -= control
                    print(f"New Volume Level: {self.volume}")
            except ValueError as e:
                print(f"Error: {e}")
            
    def set_mute(self):
        if self.power == "closed":
            print("TV is off!")
        else:
            self.volume = 0
            print(f"New Volume Level: {self.volume}")

    def num_of_channels(self):
        if self.power == "closed":
            print("TV is off!")
        else:
            a = len(channels)
            print(f"Number of channels: {a}")

    def show_channels(self):
        if self.power == "closed":
            print("TV is off!")
        else:
            x = 1
            for i in channels:
                print(f"{x}) {i}")
                x += 1
            print("\n")

    def select_random_channel(self):
        if self.power == "closed":
            print("TV is off!")
        else:
            a = random.randint(0, len(channels) - 1)
            print(f"Randomly Selected Channel: {channels[a]}")

    def add_channel(self):
        if self.power == "closed":
            print("TV is off!")
        else:
            try:
                x = int(input("How many channels do you want to add: "))
                if x < 1:
                    print("Invalid Value")
                else:
                    for i in range(x):
                        a = input(f"{i + 1}. channel you want to add: ")
                        if a in channels:
                            print("The channel is already in the list.")
                        else:
                            channels.append(a)
            except ValueError as e:
                print(f"Error: {e}")

    def surf_channels(self):
        if self.power == "closed":
            print("TV is off!")
        else:
            a = int(input("Which channel do you want to go (num): "))
            if a - 1 > len(channels):
                print("This channel is not available.")
            elif a < 1:
                print("Invalid Value")
            else:
                print(f"{channels[a - 1]} channel opened.")

    def special_apps(self):
        if self.power == "closed":
            print("TV is off!")
        else:
            for i in apps:
                print(i)
            print("\n")
            a = input("Which app do you want to open: ")
            if a not in apps:
                print("This app is not available.")
            else:
                print(f"{a} opened.")

    def commands(self):
        if self.power == "closed":
            print("TV is off!")
        else:
            try:
                print("for record: r, for play: >, for pause: =")
                a = input("Your choice: ")
                if a == "r":
                    print("Record has started.")
                    with open("tv.txt", "a", encoding="utf-8") as f:
                        f.write(f"Record Start: {datetime.now()}\n")
                elif a == ">":
                    print("Video is playing")
                    with open("tv.txt", "a", encoding="utf-8") as f:
                        f.write(f"Playing: {datetime.now()}\n")
                elif a == "=":
                    print("Video is paused")
                    with open("tv.txt", "a", encoding="utf-8") as f:
                        f.write(f"Paused: {datetime.now()}\n")
                else:
                    print("Invalid value!")
            except:
                print("Error!!!")

    def favorite_channels(self):
        if self.power == "closed":
            print("TV is off!")
        else:
            for i in fav_channels:
                print(i)
            a = input("Which channel: ")
            if a not in channels:
                print("This channel is not available.")
            else:
                if a in fav_channels:
                    print("The channel is already in the list.")
                else:
                    fav_channels.append(a)
                    print("Added.")

    def info(self):
        if self.power == "closed":
            print("TV is off!")
        else:
            print("kanal info")


remoteControl = remote_control()

print("""
Remote Control
- açma kapama (p)                   - volume (v)
- volume + (v+)                     - volume - (v-)
- mute (m)                          - show channel (ch)
- num channel (numch)               - random channel (rch)
- add channel (ch+)                 - surf channel (sch)
- special apps (app)                - record play pause (com)
- favorite channel (fav)            - info (info)
      
Çıkmak için 'q' ya basın.""")

while True:
    islem = input("işlemi seçiniz:")
    if (islem == "q"):
        print("logging out")
        break
    elif (islem == "p"):
        remoteControl.tv_power()
    elif (islem == "v"):
        remoteControl.display_volume()
    elif(islem == "v+"):
        remoteControl.increase_volume()
    elif(islem == "v-"):
        remoteControl.decrease_volume()
    elif(islem == "m"):
        remoteControl.set_mute()
    elif(islem == "ch"):
        remoteControl.show_channels()
    elif(islem == "numch"):
        remoteControl.num_of_channels()
    elif(islem == "rch"):
        remoteControl.select_random_channel()
    elif(islem == "ch+"):
        remoteControl.add_channel()
    elif(islem == "sch"):
        remoteControl.surf_channels()
    elif(islem == "app"):
        remoteControl.special_apps()
    elif(islem == "com"):
        remoteControl.commands()
    elif(islem == "fav"):
        remoteControl.favorite_channels()
    elif(islem == "info"):
        remoteControl.info()
    else:
        print("Invalid")
```
