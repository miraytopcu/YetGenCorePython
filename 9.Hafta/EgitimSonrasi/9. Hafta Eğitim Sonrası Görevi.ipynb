# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız
from datetime import datetime
import time
import random

channels = ["TRT1", "Show", "Fox", "TV8"]
apps = ["Youtube", "Netflix", "BluTV", "Amazon Prime", "Disney+", "BeIN CONNECT"]
fav_channels = []

class remote_control():
    def __init__(self, channel = 1, power = "closed", volume = 0):
        self.channel = channel
        self.power = power
        self.volume = volume

    # (Power Button) / (Channel Buttons) / (Volume Control) / (Mute) / (Numeric Keypad) / Menu / (OK-Enter) / (Directional(u-d-r-l))
    # (Back) / (Guide) / (Info) / (Special(Netflix, Youtube)) / (Record) / (Play-Pause) / (Favorite Channels)

    def tv_power(self):
        if self.power == "closed":
            print("tv açıldı")
            self.power = "opened"
        elif self.power == "opened":
            print("tv kapandı")
            self.power = "closed"

    def display_volume(self):
        if self.power == "closed":
            print("tv açık değil!")
        else:
            print(f"ses düzeyi: {self.volume}")

    def increase_volume(self):
        if self.power == "closed":
            print("tv açık değil!")
        else:
            control = int(input("sesi kaça ayarlamak istiyorsun"))
            self.volume += control
            print(f"yeni ses düzeyi {self.volume}")

    def decrease_volume(self):
        if self.power == "closed":
            print("tv açık değil!")
        else:
            control = int(input("sesi kaça ayarlamak istiyorsun"))
            self.volume -= control
            print(f"yeni ses düzeyi {self.volume}")
            
    def set_mute(self):
        if self.power == "closed":
            print("tv açık değil!")
        else:
            self.volume = 0
            print(f"yeni ses düzeyi {self.volume}")

    def num_of_channels(self):
        a = len(channels)
        print(f"kanal sayısı {a}")

    def show_channels(self):
        for i in channels:
            print(i)

    def select_random_channel(self):
        a = random.randint(0, len(channels) - 1)
        print(f"selected channel: {channels[a]}")

    def add_channel(self):
        x = int(input("kaç kanal eklemek istersin: "))
        for i in range(x):
            a = input("eklemek istediğin kanal: ")
            channels.append(a)

    def surf_channels(self):
        a = int(input("kaçıncı kanala gitmek istiyorsun:"))
        if a - 1 > len(channels):
            print("bu kanal mevcut değil")
        else:
            print(f"{channels[a - 1]} kanalına geçildi.")

    def special_apps(self):
        for i in apps:
            print(i)
        a = input("hangisini açmak istiyorsun")
        print(f"{a} açıldı")

    def commands(self):
        print("record- r, play- >, pause- =")
        a = input("tercih")
        if a == "r":
            print("kayıt başladı")
        if a == ">":
            print("oynatılıyor")
        if a == "=":
            print("durduruldu")
        else:
            print("işlem geçersiz")

    def favorite_channels(self):
        for i in fav_channels:
            print(i)
        a = input("hangi kanalı")
        if a not in channels:
            print("böyle bir kanal mevcut değil")
        else:
            if a in fav_channels:
                print("zaten mevcut")
            else:
                fav_channels.append(a)
                print("eklendi")

    def info(self):
        print("kanal info")

    def menu(self):
        print("menu")

remoteControl = remote_control()

print("""
Remote Control
- açma kapama (p)
- volume (v)
- volume + (v+)
- volume - (v-)
- mute (m)
- show channel (ch)
- num channel (numch)
- random channel (rch)
- add channel (ch+)
- surf channel (sch)
- special apps (app)
- record play pause (com)
- favorite channel (fav)
- info (info)
- menu (menu)
Çıkmak için 'q' ya basın.""")

while True:
    islem = input("işlemi seçiniz:")
    if (islem == "q"):
        print("sistemden çıkılıyor")
        break
    elif (islem == "p"):
        remoteControl.tv_power()
    elif (islem == "v"):
        remoteControl.display_volume()
    elif(islem == "v+"):
        remoteControl.increase_volume()
    elif(islem == "v-"):
        remoteControl.decrease_volume()
    elif(islem == "m"):
        remoteControl.set_mute()
    elif(islem == "ch"):
        remoteControl.show_channels()
    elif(islem == "numch"):
        remoteControl.num_of_channels()
    elif(islem == "rch"):
        remoteControl.select_random_channel()
    elif(islem == "ch+"):
        remoteControl.add_channel()
    elif(islem == "sch"):
        remoteControl.surf_channels()
    elif(islem == "app"):
        remoteControl.special_apps()
    elif(islem == "com"):
        remoteControl.commands()
    elif(islem == "fav"):
        remoteControl.favorite_channels()
    elif(islem == "info"):
        remoteControl.info()
    elif(islem == "menu"):
        remoteControl.menu()
    else:
        print("geçersiz")
```
